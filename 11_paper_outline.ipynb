## ABSTRACT (150–200 words)

**Template:**

> We apply **ordinal network analysis** to investigate the price dynamics and market efficiency of eight major cryptocurrencies (BTC, ETH, BNB, SOL, XRP, ADA, DOGE, LTC) over a six-year period (April 2020 – April 2026) comprising **2,210 daily observations**. By encoding log-return time series into ordinal transition networks, we characterize market structure through **permutation entropy (PE)** and the **spectral gap** of transition matrices. Our analysis reveals that all eight markets operate near maximum entropy (PE ≈ 0.9997 ± 0.0003), with surrogate data tests confirming that no cryptocurrency exhibits detectable ordinal structure beyond phase-randomized benchmarks (*p* > 0.05 for all coins). Time-varying PE analysis identifies transient efficiency losses around major market events — notably following the LUNA collapse (May 2022) and the 2025 correction — but these effects are statistically small (|ΔPE| < 0.008). Regime-specific Kruskal-Wallis tests show significant PE variation across five market regimes for **7/8 coins** (*p* < 0.001), though effect sizes remain small (rank-biserial *r* < 0.16, Bonferroni-corrected). Walk-forward prediction experiments confirm that ordinal-based classifiers achieve accuracy indistinguishable from chance (48–52%), consistent with weak-form efficiency. These findings suggest that while ordinal network topology provides a rich characterization of cryptocurrency market microstructure, the high entropy of these markets severely limits short-term return predictability.

**Keywords:** ordinal networks, permutation entropy, cryptocurrency, market efficiency, efficient market hypothesis, regime analysis, surrogate data test

## 3. DATA AND METHODOLOGY (~800 words)

### 3.1 Data
- **Source:** CryptoCompare API (daily OHLCV)  
- **Assets:** BTC, ETH, BNB, SOL, XRP, ADA, DOGE, LTC — top-8 by market cap as of 2024  
- **Period:** 2020-04-10 → 2026-04-29 (**2,210 trading days**)  
- **Variable:** Log-returns $r_t = \ln(P_t / P_{t-1})$  
- **Stationarity:** ADF tests reject unit root for all 8 series (*p* < 0.001)  
- **Cleaning:** 0 missing values after balanced panel construction  

### 3.2 Ordinal Pattern Encoding
Given embedding dimension $d$ and delay $\tau$, an **ordinal pattern** $\pi_t$ is the rank-order of the subsequence $(r_t, r_{t+\tau}, \ldots, r_{t+(d-1)\tau})$:
$$\pi_t = \text{argsort}\bigl(\text{argsort}(r_t, r_{t+\tau}, \ldots, r_{t+(d-1)\tau})\bigr)$$
Parameters selected by grid search: **$d^* = 3$, $\tau^* = 1$** (minimizing walk-forward forecast error across all coins).

### 3.3 Ordinal Network Construction
The **ordinal transition network** $G = (V, E, W)$ connects pattern $\pi_i \to \pi_j$ with weight proportional to transition frequency:
$$T_{ij} = \frac{\#(\pi_t = \pi_i, \pi_{t+1} = \pi_j)}{\#(\pi_t = \pi_i)}$$
Network metrics extracted:
- **Permutation Entropy:** $H(d) = -\sum p(\pi) \ln p(\pi) / \ln(d!)$  
- **Spectral gap:** $\lambda_1 - \lambda_2$ of transition matrix $T$ (speed of Markov chain mixing)  
- **Forbidden patterns:** patterns with $p(\pi) = 0$ (indicate deterministic structure)

### 3.4 Robustness Parameters
- **Embedding dimension:** $d \in \{3, 4, 5\}$  
- **Delay:** $\tau \in \{1, 2, 3, 5\}$  
- **Scale:** Multiscale PE at $s \in \{1, 2, 3, 5, 7, 10\}$ days

## 5. DISCUSSION (~500 words)

### 5.1 Implications for Market Efficiency
- Consistent with **Adaptive Market Hypothesis** (Lo, 2004): efficiency is not binary but time-varying
- Near-maximal PE ≈ 0.9997 suggests crypto log-returns approach the theoretical maximum disorder
- Lack of forbidden patterns at d=3,4,5 → no deterministic structure, fully stochastic transitions

### 5.2 What Ordinal Networks Add Beyond PE
- **Spectral gap** distinguishes mixing speed (BTC: 0.512 vs SOL: 0.509) — may predict liquidity recovery time
- **Transition matrix topology** reveals which pattern sequences are over/under-represented (even when marginal PE ≈ max)
- Framework is naturally **regime-interpretable**: transition matrix changes between Bear and Bull are measurable

### 5.3 Limitations
1. **Daily data only** — intraday structure (e.g., 1-hour) may show different patterns  
2. **d=3** gives only 6 patterns — may be insufficient to detect subtle structure (d=4,5 more sensitive but require longer series)  
3. **Effect sizes small** — statistical significance does not imply trading opportunity after transaction costs  
4. **Regime labels manually defined** — alternative: use regime-switching model (Markov HMM) for endogenous regimes  

### 5.4 Comparison with Existing Literature
- Bariviera et al. (2017): Bitcoin Hurst exponent → long-range dependence. Our result: PE ≈ 1 → *no* such dependence at daily scale after 2020  
- Zhang et al. (2021): multifractal → crypto less efficient than stocks. Our result: PE same order as theoretical maximum → more nuanced picture  
- *Key claim:* Ordinal network framework is more informative than standalone PE because it captures second-order transition structure

## 6. CONCLUSION (~300 words)

> This study applies ordinal network analysis to characterize the price dynamics and predictability of eight major cryptocurrencies from 2020 to 2026. Using permutation entropy and spectral gap as primary metrics, we find that all markets operate near theoretical maximum entropy, with no coin exhibiting ordinal structure distinguishable from phase-randomized surrogates. These findings strongly support weak-form market efficiency at the daily time scale.
>
> Time-varying analysis reveals that efficiency is not constant: transient decreases in PE are associated with major market stress events (notably the 2025 correction, ΔPE = −0.0046), while post-shock efficiency recovery is rapid (within 30 days). Regime-stratified analysis confirms statistically significant PE differences across five market cycles (Kruskal-Wallis *p* < 0.001 for 7/8 coins), though effect sizes remain small, reinforcing the practical ineffectiveness of ordinal-based trading strategies.
>

>
> Future work should extend this framework to intraday data, incorporate on-chain metrics as external covariates, and explore whether ordinal network topology can serve as a systemic risk indicator during market crises.
