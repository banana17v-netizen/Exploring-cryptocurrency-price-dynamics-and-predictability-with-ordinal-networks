## TITLE PAGE

**Title:** Exploring Cryptocurrency Price Dynamics and Predictability with Ordinal Networks

**Abstract**

We apply ordinal network analysis to investigate the price dynamics and market efficiency of eight major cryptocurrencies (BTC, ETH, BNB, SOL, XRP, ADA, DOGE, LTC) over a six-year period spanning April 2020 to April 2026, comprising 2,210 daily log-return observations. By encoding return time series into ordinal transition networks, we characterize market microstructure through permutation entropy (PE) and the spectral gap of transition matrices. All eight markets exhibit near-maximal entropy (PE ∈ [0.9995, 0.9999]; mean = 0.9997 ± 0.0002), and surrogate data tests confirm that no cryptocurrency exhibits ordinal structure distinguishable from phase-randomized benchmarks (z-scores ∈ [−1.14, +0.24]; p > 0.05 for all coins). Regime-stratified Kruskal-Wallis tests reveal statistically significant PE variation across five market cycles for 7/8 coins (p < 0.001), with Ethereum (ETH) representing a notable exception consistent with superior market integration (H = 3.43, p = 0.489). Time-varying PE analysis identifies transient efficiency reductions around major stress events, including the 2025 correction (ΔPE = −0.0046), though recovery occurs within 30 days. Walk-forward prediction experiments using ordinal-based classifiers (Logistic Regression, Random Forest, Gradient Boosting, MLP) confirm accuracy indistinguishable from chance (48–52%), with no model significantly outperforming ARIMA(1,0,1), GARCH(1,1), LSTM, or Transformer deep learning baselines after Bonferroni-corrected Diebold-Mariano testing. These results support weak-form market efficiency at the daily scale, while demonstrating that ordinal network topology provides a richer characterization of market microstructure than scalar entropy measures alone.

**Keywords:** ordinal networks; permutation entropy; cryptocurrency; market efficiency; efficient market hypothesis; regime analysis; surrogate data test; spectral gap

## 1. INTRODUCTION

Cryptocurrency markets have grown from a niche technological curiosity into a global asset class with combined market capitalizations exceeding one trillion US dollars. Unlike traditional financial markets, cryptocurrency exchanges operate continuously — 24 hours a day, 7 days a week — and are characterized by high volatility, fragmented liquidity, and diverse participant bases ranging from retail investors to algorithmic trading desks. These structural features raise fundamental questions about the informational efficiency of cryptocurrency prices and the extent to which past returns contain predictive information about future price movements.

The Efficient Market Hypothesis (EMH) in its weak form posits that current asset prices fully reflect all information contained in the history of past prices, rendering technical or pattern-based forecasting strategies unprofitable on a risk-adjusted basis [7]. Empirical evidence for weak-form efficiency in cryptocurrency markets is mixed: early studies found significant inefficiencies in Bitcoin [1], while more recent analyses suggest that efficiency has evolved over time and may vary across market conditions [3,4].

Traditional approaches to testing market efficiency — autocorrelation tests, variance ratio tests, or GARCH-based volatility modeling — are largely linear and may fail to detect nonlinear dependencies or deterministic structure in return dynamics. Ordinal pattern analysis, introduced by Bandt and Pompe [8], offers a nonparametric alternative that is robust to noise, requires no distributional assumptions, and captures the rank-order structure of time series at arbitrary embedding scales. Permutation entropy (PE), derived from ordinal patterns, has been applied to stock markets [10] and financial time series broadly [9], but its extension to ordinal *networks* — which capture second-order transition structure between successive patterns — has received limited attention in the context of cryptocurrency markets.

The ordinal partition network framework, formalized by Small [12] and extended by McCullough et al. [13] and Pessa and Ribeiro [14], constructs a directed weighted graph where nodes represent distinct ordinal patterns and edges represent empirical transition probabilities. This representation encodes not only the marginal distribution of patterns (captured by PE) but also their sequential dependencies — information lost when computing scalar entropy alone. The spectral gap of the transition matrix, for instance, characterizes the speed of Markov chain mixing and thus the rate at which market memory decays.

This paper makes the following contributions. First, we conduct the most comprehensive ordinal network analysis of cryptocurrency markets to date, covering eight major coins across a six-year period (2020–2026) that encompasses five distinct market regimes, including the COVID-19 recovery, the LUNA ecosystem collapse, the FTX bankruptcy, the Bitcoin ETF approval cycle, and the 2025 macroeconomic correction. Second, we integrate surrogate data testing, block bootstrap confidence intervals, multiscale permutation entropy, and Bonferroni-corrected regime comparisons into a unified methodological framework, enabling rigorous statistical inference about market efficiency. Third, we demonstrate that Ethereum exhibits a qualitatively distinct pattern of near-constant efficiency across all market regimes — a finding not previously reported — which we interpret as evidence of superior market integration relative to other major cryptocurrencies. Fourth, through walk-forward prediction experiments with multiple ordinal classifiers, we provide strong negative evidence against ordinal predictability at the daily scale, contributing to the growing literature on the practical limits of technical analysis in modern cryptocurrency markets.

The remainder of this paper is organized as follows. Section 2 reviews the relevant literature on cryptocurrency market efficiency and ordinal network methods. Section 3 describes the data and methodology. Section 4 presents our empirical results. Section 5 discusses implications and limitations. Section 6 concludes.

## 2. LITERATURE REVIEW

### 2.1 Market Efficiency in Cryptocurrency Markets

The question of informational efficiency in cryptocurrency markets has generated a substantial empirical literature since Bitcoin's emergence as a tradable asset. Urquhart (2016) applied Ljung-Box autocorrelation tests and variance ratio tests to Bitcoin daily returns, finding significant predictability in early data that weakened substantially over the 2011–2016 period. Nadarajah and Chu (2017) reanalyzed the same data using power transformations and found evidence consistent with weak-form efficiency. Bariviera et al. (2017) applied detrended fluctuation analysis to compute Hurst exponents, documenting time-varying long-range dependence in Bitcoin returns — a finding consistent with the Adaptive Market Hypothesis (AMH) proposed by Lo (2004), which views efficiency as an evolving property shaped by institutional changes, participant learning, and market stress.

Tran and Leirvik (2020) provided direct tests of the AMH in cryptocurrency markets, finding that efficiency measures fluctuate across time and differ across coins, with Bitcoin generally more efficient than smaller-cap alternatives. Zhang et al. (2021) applied multifractal detrended fluctuation analysis to a panel of cryptocurrencies, concluding that crypto markets are less efficient than traditional equity markets but more efficient than earlier studies suggested — a finding attributed to growing institutional participation and improved market microstructure.

A limitation of these studies is their reliance on linear or fractal-scaling measures that may not capture discrete nonlinear structure. Our analysis addresses this by employing ordinal network methods, which are sensitive to rank-order dependencies at arbitrary time scales.

### 2.2 Ordinal Patterns and Permutation Entropy

Bandt and Pompe (2002) introduced permutation entropy as a simple, robust complexity measure for time series. For an embedding dimension d, PE is computed from the probability distribution of the d! possible rank-order patterns (ordinal patterns) within consecutive length-d subsequences. PE = 1 implies all patterns are equally probable (maximum disorder); PE < 1 indicates distributional asymmetry consistent with deterministic or autocorrelated structure.

Zunino et al. (2010) applied PE to compare market efficiency across global stock indices using the complexity–entropy causality plane, finding that developing markets exhibit lower PE than developed markets. Zanin et al. (2012) extended this approach to intraday financial data and documented PE dependencies on time resolution. Olivares et al. (2012) introduced the concept of forbidden ordinal patterns — patterns with zero empirical probability — as a diagnostic for deterministic dynamics, finding that truly stochastic processes exhibit no forbidden patterns.

### 2.3 Ordinal Networks

The transition from scalar PE to network representations was formalized by Small (2013), who constructed ordinal partition networks by connecting successive patterns as nodes in a directed graph. McCullough et al. (2015) demonstrated that network topology metrics — degree distribution, clustering coefficient, spectral properties — provide discrimination between chaotic and stochastic processes beyond what scalar PE can offer. Pessa and Ribeiro (2019) provided a comprehensive theoretical treatment, showing that the spectral gap of the transition matrix governs the rate of information decay (Markov mixing time) and correlates with the predictive horizon of the underlying dynamics.

To our knowledge, no prior study has applied the full ordinal network framework — including transition matrix spectral analysis, network centrality measures, and regime-stratified comparisons — to a comprehensive panel of cryptocurrency markets. This represents the primary methodological contribution of the present work.

## 3. DATA AND METHODOLOGY

### 3.1 Data

We use daily closing prices for eight major cryptocurrencies — Bitcoin (BTC), Ethereum (ETH), Binance Coin (BNB), Solana (SOL), XRP, Cardano (ADA), Dogecoin (DOGE), and Litecoin (LTC) — sourced from the CryptoCompare API. These eight assets represent the top cryptocurrencies by market capitalization as of 2024 and collectively account for the majority of total cryptocurrency market capitalization. The sample period runs from April 10, 2020 to April 29, 2026, yielding **T = 2,210 daily observations** after constructing a balanced panel with no missing values. The sample encompasses five distinct market regimes: the COVID-19 recovery and 2020–2021 bull market, the 2022 bear market (including the LUNA/Terra collapse in May 2022 and the FTX bankruptcy in November 2022), the 2023 recovery period, the 2024–2025 bull market (following Bitcoin ETF approval in January 2024), and the 2025 macroeconomic correction.

The primary return series is the daily log-return:
$$r_t = \ln(P_t / P_{t-1})$$
where $P_t$ is the closing price on day $t$. Augmented Dickey-Fuller (ADF) tests reject the unit root null hypothesis for all eight series at the 0.1% significance level, confirming stationarity of the return series.

### 3.2 Ordinal Pattern Encoding

Given embedding dimension $d$ and time delay $\tau$, the ordinal pattern $\pi_t \in S_d$ (the symmetric group of permutations of order $d$) is defined as the rank-order permutation of the subsequence $(r_t, r_{t+\tau}, \ldots, r_{t+(d-1)\tau})$:
$$\pi_t = \text{argsort}\bigl(\text{argsort}(r_t, r_{t+\tau}, \ldots, r_{t+(d-1)\tau})\bigr)$$
This encoding yields $d!$ possible patterns. For $d = 3$, there are $3! = 6$ patterns; for $d = 4$ and $d = 5$, there are 24 and 120 patterns, respectively.

Optimal parameters are selected by grid search over $d \in \{3, 4, 5\}$ and $\tau \in \{1, 2, 3, 5\}$, minimizing walk-forward directional forecast error across all eight coins. The consensus optimal parameters are $d^* = 3$, $\tau^* = 1$, which we adopt as baseline; robustness to alternative parameter choices is reported in Section 4.6.

### 3.3 Ordinal Transition Network

The ordinal transition network $G = (V, E, W)$ is a directed weighted graph with node set $V = \{\pi_1, \ldots, \pi_{d!}\}$ (the $d!$ ordinal patterns) and edge weight $W_{ij}$ equal to the empirical transition probability from pattern $\pi_i$ to $\pi_j$:
$$T_{ij} = P(\pi_{t+1} = \pi_j \mid \pi_t = \pi_i) = \frac{\#\{t : \pi_t = \pi_i, \pi_{t+1} = \pi_j\}}{\#\{t : \pi_t = \pi_i\}}$$
The matrix $T = (T_{ij})$ is row-stochastic (i.e., $\sum_j T_{ij} = 1$) and represents a Markov chain over the ordinal pattern space.

We extract the following network metrics:
- **Permutation entropy:** $H(d) = -\sum_{\pi} p(\pi) \ln p(\pi) \,/\, \ln(d!)$, where $p(\pi)$ is the marginal frequency of pattern $\pi$. $H = 1$ denotes maximum disorder.
- **Spectral gap:** $\Delta\lambda = \lambda_1 - \lambda_2$, where $\lambda_1 = 1 \geq \lambda_2 \geq \ldots$ are the eigenvalues of $T$ sorted in descending order. The spectral gap governs the Markov mixing time: larger $\Delta\lambda$ implies faster decay of pattern memory.
- **Forbidden patterns:** $n_f = |\{\pi : p(\pi) = 0\}|$. A non-zero count indicates deterministic structure.
- **Network centrality:** PageRank, degree centrality, and betweenness centrality of pattern nodes, characterizing the topology of pattern transition flows.

### 3.4 Statistical Inference

**Surrogate data test.** To test whether observed PE is distinguishable from a phase-randomized process with identical spectral properties, we generate $N = 500$ surrogate time series per coin using the Iterative Amplitude Adjusted Fourier Transform (IAAFT) algorithm [16]. The z-score $z = (PE_{\text{real}} - \bar{PE}_{\text{surr}}) / \sigma_{\text{surr}}$ follows an approximately normal distribution under the null hypothesis of no ordinal structure beyond the linear autocorrelation spectrum.

**Bootstrap confidence intervals.** We compute 95% confidence intervals for PE using block bootstrap resampling ($N = 1{,}000$ resamples, block size $b = 20$ days) to account for temporal dependence in the return series [17].

**Regime comparison.** We define five market regimes based on key structural events (Table 5) and test for PE differences across regimes using the nonparametric Kruskal-Wallis H-test. Pairwise comparisons are conducted using the Mann-Whitney U-test with Bonferroni correction for multiple comparisons (10 pairs per coin); Benjamini-Hochberg false discovery rate (FDR) correction is reported as supplementary analysis. Effect sizes are quantified using the rank-biserial correlation $r_{\text{rb}} \in [-1, 1]$.

**Predictability test.** Walk-forward (expanding window) prediction experiments are conducted with a training window of 365 days and test window of 30 days, stepping forward 30 days at a time. Four ordinal classifiers are evaluated: Logistic Regression (LogReg), Random Forest (RF), Gradient Boosting Machine (GBM), and Multilayer Perceptron (MLP), all using the encoded ordinal pattern sequence as the sole input feature. Six baseline models are evaluated: ARIMA(1,0,1), GARCH(1,1) directional forecast, naïve persistence, naïve majority vote, random walk, and two deep learning architectures — a two-layer LSTM (hidden=32, seq\_len=20, epochs=10) and a single-encoder Transformer (embed\_dim=8, nhead=2, seq\_len=20, epochs=10) — both operating on the same ordinal pattern sequence. All baseline models were pre-specified based on prior literature before ordinal classifier results were evaluated, in accordance with standard forecast evaluation protocol (Diebold & Mariano, 1995). Predictive performance is compared using the Diebold-Mariano test [18] with Newey-West standard errors and Bonferroni correction for multiple simultaneous comparisons.

Several features are noteworthy. First, all PE values are tightly clustered in the interval [0.9995, 0.9999], indicating near-maximal ordinal disorder across all coins and the full sample. The cross-sectional variation in spectral gap, while small in absolute magnitude, is interpretable: BTC's larger spectral gap (0.534) implies faster decay of pattern autocorrelation compared to ETH (0.491) and LTC (0.495), consistent with BTC's dominant market share and price discovery role.

### 4.3 Time-Varying Market Efficiency

Despite the high average PE, a rolling analysis (window = 60 days, step = 5 days) reveals meaningful temporal variation. The cross-sectional mean PE ranges from a minimum of 0.863 to a maximum of 0.999 over the sample period (Figure 2). Several key episodes are identifiable:

- **Post-LUNA collapse (May 2022):** Mean PE increased by +0.0051 in the 30 days following the LUNA/Terra ecosystem collapse, suggesting a counterintuitive *improvement* in ordinal efficiency immediately post-event. We interpret this as reflecting rapid information absorption — the shock was so severe that return patterns became more random as prices adjusted to a new equilibrium under high uncertainty.
- **2025 correction (April 2025):** Mean PE decreased by −0.0046, representing the most persistent efficiency reduction in the sample. Unlike the 2022 events, the 2025 correction was driven by macroeconomic factors (Federal Reserve policy reversal, global risk-off) rather than crypto-specific shocks, potentially explaining the slower ordinal structure response.
- **Bull markets:** Both bull market regimes (2020–21 and 2024–25) show elevated PE, consistent with strong trend-following and information-efficient price discovery during period of broad market participation.

Figure 3 presents a heatmap of per-coin rolling PE, revealing substantial cross-sectional dispersion during the Bear 2022 regime — coins diverge in their efficiency trajectory during systemic stress events.

### 4.6 Robustness Checks

**Parameter sensitivity (τ variation).** Computing PE at $d = 3$ across $\tau \in \{1, 2, 3, 5\}$, the coefficient of variation (CV) of PE across delay values is below 0.035% for all eight coins, confirming that results are insensitive to the choice of embedding delay.

**Embedding dimension variation.** At $d = 4$ (24 patterns), mean PE across all coins is 0.9984; at $d = 5$ (120 patterns), mean PE is 0.9942. While slightly more structure is detectable at larger $d$ — consistent with the known property that higher-order patterns can capture more subtle dependencies — PE remains near-maximal and the qualitative conclusions are unchanged.

**Multiscale permutation entropy.** Applying coarse-graining at scale $s \in \{1, 2, 3, 5, 7, 10\}$ days, PE decreases monotonically from 0.9997 (scale 1) to 0.9958 (scale 10). The slight PE reduction at coarser scales is consistent with the weak Ljung-Box autocorrelation detected at lags of 5–10 days and likely reflects the influence of weekly trading patterns. This finding suggests that while daily returns exhibit maximum ordinal disorder, lower-frequency return dynamics contain marginally more predictable structure.

**Cross-coin lead-lag analysis.** To test whether BTC return patterns systematically lead those of altcoins — a hypothesis consistent with BTC's price-discovery role — we compute cross-correlation of rolling PE time series between BTC and each altcoin at lags $\ell \in \{-5, \ldots, +5\}$ days. For all seven altcoins, the maximum cross-correlation occurs at lag $\ell = 0$ (contemporaneous), with lag-0 correlations ranging from $r = 0.539$ (SOL) to $r = 0.686$ (ETH). Neither BTC leads any altcoin nor do altcoins lead BTC at any lag examined. This absence of lead-lag structure at the ordinal PE level is consistent with the efficient market hypothesis and contradicts the notion of a hierarchical information cascade from BTC to altcoins at daily frequency.

## 5. DISCUSSION

### 5.1 Implications for Market Efficiency

Our results provide strong, methodology-independent support for weak-form efficiency in major cryptocurrency markets at the daily time scale. The near-maximal PE (mean = 0.9997), the complete absence of forbidden ordinal patterns, and the failure of surrogate tests to detect structure beyond linear dependence collectively suggest that daily cryptocurrency log-returns behave as ordinal random walks — generating all rank-order patterns with approximately equal probability and without systematic sequential dependencies.

This finding contrasts with earlier studies documenting significant inefficiency in Bitcoin (Urquhart 2016) and long-range dependence (Bariviera et al. 2017), but is consistent with the temporal evolution hypothesis: as cryptocurrency markets have matured, deepened, and attracted institutional participation, the degree of short-term predictability has converged toward zero. Our six-year sample, which begins in April 2020 — well after the market's early, thinly-traded phase — reflects this mature market structure.

The time-varying PE analysis supports Lo's (2004) Adaptive Market Hypothesis: efficiency fluctuates in response to structural shocks, but recovers quickly (within 30 days) after endogenous crypto-specific events. The more persistent efficiency loss observed during the 2025 macroeconomic correction may reflect the growing integration of cryptocurrency markets with traditional financial markets and their susceptibility to common macro risk factors.

### 5.2 What Ordinal Networks Add Beyond Scalar Entropy

A key contribution of the ordinal network framework is that it provides richer market characterization than PE alone. Three findings illustrate this:

First, the **spectral gap** varies meaningfully across coins (0.491 for ETH vs. 0.534 for BTC) despite nearly identical PE values. This variation characterizes the Markov mixing time — the speed at which ordinal pattern memory decays — and is not recoverable from scalar PE. BTC's larger spectral gap implies faster information processing relative to ETH, consistent with its larger share of overall price discovery in the cryptocurrency ecosystem.

Second, **regime-conditional transition matrices** reveal how the relative probabilities of pattern transitions change between bull and bear markets, even when average PE remains near maximal. This second-order characterization is invisible to univariate PE analysis.

Third, the **ETH regime-invariance finding** — ETH is the only coin whose PE does not vary significantly across five market regimes — emerged from the regime-stratified network analysis and would not have been detectable through cross-sectional PE comparison alone.

### 5.3 Limitations

Several limitations should be noted. First, our analysis is restricted to daily closing prices; intraday ordinal network analysis (e.g., at hourly resolution) may reveal different dynamics, particularly around high-frequency events such as exchange outages or large liquidation cascades. Second, the choice of $d = 3$ yields only six ordinal patterns, which may be insufficient to detect subtle higher-order structure; the slightly lower PE at $d = 5$ suggests that extending to larger pattern spaces is a worthwhile direction. Third, regime boundaries are defined exogenously based on qualitative criteria; an endogenous regime identification approach (e.g., hidden Markov models) could yield different regime compositions and potentially sharper PE contrasts. Fourth, transaction costs, slippage, and funding rates — ignored in our Sharpe ratio calculations — would further erode the practical value of any ordinal-based trading strategy.

### 5.4 Directions for Future Research

Future work could extend the ordinal network framework in several directions: (i) applying the method to intraday data, where structural dependencies are more likely to persist; (ii) incorporating on-chain blockchain metrics (transaction volume, active addresses, exchange inflows) as covariates to explain PE dynamics; (iii) investigating whether the spectral gap serves as an early warning indicator for liquidity crises or regime transitions; and (iv) extending the cross-coin analysis to examine how the joint ordinal network of multiple cryptocurrencies encodes systemic risk.

## 6. CONCLUSION

This study applies ordinal network analysis to characterize the price dynamics and return predictability of eight major cryptocurrencies across a six-year period (April 2020 – April 2026) encompassing five distinct market regimes and 2,210 daily observations. Using permutation entropy and the spectral gap of ordinal transition matrices as primary metrics, supplemented by surrogate data testing, block bootstrap confidence intervals, multiscale analysis, and regime-stratified nonparametric tests, we arrive at four principal conclusions.

First, all eight cryptocurrencies exhibit near-maximal ordinal entropy (mean PE = 0.9997) with no forbidden patterns, and real PE is statistically indistinguishable from phase-randomized surrogates (all z-scores ∈ [−1.14, +0.24]). These findings support weak-form market efficiency at the daily scale.

Second, efficiency is not stationary: time-varying PE reveals transient reductions following major market stress events, most prominently the 2025 macroeconomic correction (ΔPE = −0.0046), with recovery occurring within approximately 30 days for crypto-specific shocks. This is consistent with the Adaptive Market Hypothesis.

Third, Ethereum exhibits a qualitatively distinct regime-invariant efficiency profile — the only coin whose PE does not vary significantly across five market regimes (KW: H = 3.43, p = 0.489) — which we attribute to its superior liquidity and market integration. This finding provides new evidence for heterogeneity in efficiency levels across the cryptocurrency cross-section.

Fourth, walk-forward prediction experiments confirm that ordinal-based classifiers achieve accuracy statistically indistinguishable from ARIMA(1,0,1), GARCH(1,1), naïve baselines, LSTM, and Transformer deep learning architectures for all eight coins after Bonferroni-corrected Diebold-Mariano testing. This negative result reinforces the practical limits of pattern-based and deep learning approaches in mature daily cryptocurrency markets.

Together, these findings demonstrate that while ordinal network topology provides a richer characterization of cryptocurrency market microstructure than scalar entropy measures alone, the high ordinal entropy of daily returns severely constrains short-term predictability and is inconsistent with economically meaningful inefficiency.
